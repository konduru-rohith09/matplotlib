**Matplotlib For Machine learning**

Matplotlib is a Python library used to create graphs and visualizations.

It helps you see your data instead of just looking at numbers.

**Line Plot**

This graph shows how the training loss changes over multiple epochs during model training.

Epochs (x-axis): Number of times the model has seen the full training dataset

Loss (y-axis): How badly the model is performing (lower = better)


What the curve shows:

At the beginning (Epoch 1), the loss is high (0.9) because the model is still learning.

As training continues, the loss keeps decreasing.

By Epoch 5, the loss becomes 0.25, which means the model has improved significantly.

```python
import matplotlib.pyplot as plt # importing matplotlib

epochs = [1, 2, 3, 4, 5] # Training iteration numbers
loss = [0.9, 0.6, 0.4, 0.3, 0.25] # loss values for each epochs

plt.plot(epochs, loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.show()
```

**Scatter plot**

This scatter plot shows the relationship between the input feature (x-axis) and the target/output value (y-axis).

What the plot shows

Each dot represents one data sample

The x-value is the input feature

The y-value is the corresponding target value

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5] # feature values
y = [2, 4, 5, 4, 5] # target values

plt.scatter(x, y)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Feature vs Target")
plt.grid(True)
plt.show()
```

**Histogram**

This histogram shows the distribution of a numerical feature in the dataset.

The data is generated from a normal (Gaussian) distribution with:

Mean = 50

Standard Deviation = 10

Total Samples = 200

What the plot shows

The x-axis represents the value range of the feature

The y-axis represents how many data points fall within each range (frequency)

The bars show how often values occur

bins=20 splits the data into 20 value groups

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(50, 10, 200)

plt.hist(data, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Feature Distribution")
plt.grid(True)
plt.show()
```

**Bar chart**

This bar chart shows how many samples belong to each class in the dataset.

What the plot shows

Class 0 contains 120 samples

Class 1 contains 80 samples

The x-axis represents the class labels, and the y-axis represents the number of data points in each class.




```python
import matplotlib.pyplot as plt

classes = ["Class 0", "Class 1"]
count = [120, 80]

plt.bar(classes, count)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Class Distribution")
plt.grid(True)
plt.show()
```

**Box plot**

This box plot is used to visualize the spread of the data and identify outliers.

A box plot displays:

Median (middle value)

Interquartile Range (IQR: middle 50% of data)

Whiskers (normal data range)

Outliers (points far from the rest)

```python
import matplotlib.pyplot as plt
import numpy as np

data = [10, 12, 14, 15, 18, 50]

plt.boxplot(data)
plt.title("Box Plot - Outlier Detection")
plt.grid(True)
plt.show()
```

**Subplots**
This figure shows two separate line plots:
one for Training Loss and one for Validation Loss across multiple epochs.

What the plots represent
 Training Loss (Left Plot)

Measures how well the model fits the training data  the data it learns from.

Starts high at 0.9

Gradually decreases to 0.25

This means the model is learning and improving over time.


Validation Loss (Right Plot)

Measures performance on unseen validation data  data not used for training.

Starts at 1.0

Decreases to 0.35

This shows the model is also improving on data it hasnt seen before.

```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.6, 0.4, 0.3, 0.25]
val_loss = [1.0, 0.7, 0.5, 0.4, 0.35]

fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(epochs, train_loss)
ax[0].set_title("Training Loss")

ax[1].plot(epochs, val_loss)
ax[1].set_title("Validation Loss")

plt.show()
```

**confusion matrix**

This confusion matrix compares the actual class labels with the model-predicted labels for a binary classification problem.


The heatmap displays a 22 matrix.

Meaning of Each Value

True Positive (TP = 2)
Model correctly predicted class 1

True Negative (TN = 2)
Model correctly predicted class 0

False Positive (FP = 1)
Model predicted 1, but actual was 0

False Negative (FN = 1)
Model predicted 0, but actual was 1

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_true = [0, 0, 1, 1, 0, 1]
y_pred = [0, 1, 1, 1, 0, 0]

cm = confusion_matrix(y_true, y_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

**ROC curve**

This plot visualizes the Receiver Operating Characteristic (ROC) Curve, which is used to evaluate the performance of a binary classification model.

What the ROC Curve Shows

The ROC curve plots:

False Positive Rate (FPR) on the x-axis

True Positive Rate (TPR / Recall) on the y-axis

Each point represents the model performance at a different probability threshold.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

y_true = [0, 0, 1, 1]
y_prob = [0.1, 0.4, 0.35, 0.8]

fpr, tpr, _ = roc_curve(y_true, y_prob)
auc = roc_auc_score(y_true, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```
